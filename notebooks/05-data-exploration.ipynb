```python
import pandas as pd
import altair as alt
```

```python
data = pd.read_csv("../data/processed/val/data.csv")
```

```python
data.head()
```

```python
file = data["file"][4]
```

```python
base = alt.Chart(data[data["file"]==file])

superflux = base.mark_line().encode(x="timestamps", y="superflux")

onsets = alt.Chart(data[data["file"]==file][data["onset"]]).mark_rule(color='red').encode(
    x='timestamps'
)

(superflux + onsets).properties(width=1200).interactive()
```

```python
base = alt.Chart(data[data["file"]==file])

features = base.mark_line().encode(x="timestamps", y=alt.Y(alt.repeat("row"), type='quantitative')).properties(width=1200)

onsets = alt.Chart(data[data["file"]==file][data["onset"]]).mark_rule(color='red').encode(
    x='timestamps'
)

(features + onsets).repeat(row=["superflux", "welch", "kurtosis"])
```

```python
alt.Chart(data.head(100)).mark_circle(opacity=0.5).encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='onset:N'
).properties(
    width=100,
    height=100
).repeat(
    row=["superflux", "energy", "welch", "kurtosis", "skew"],
    column=list(reversed(["superflux", "energy", "welch", "kurtosis", "skew"]))
).interactive()
```
